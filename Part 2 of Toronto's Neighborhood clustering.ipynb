```python
# shorten the name of the dataframe above by assigning dff<----df_data_1
dff = df_data_1
dlist = dff['Postal Code'].to_list()
print(dlist)
```

### Next is to generate corresponding latitudes and longitudes to the postal codes listed above using the pgeocode Nominatim

```python
ca_1 = pgeocode.Nominatim('ca')
ca_2 = ca_1.query_postal_code(dlist)
ca_2
```

### In the above dataframe, what are needed here are the Latitudes and Longitudes of the Postal Codes of the State of     Ontario. So every other attributes or columns have to be dropped from the dataframe.

### It is also observed that row #76 does not have latitude and longitude, in other words they were not available. However, we acquired the latitude and longitude for this Borough which we subsequently added to replace the NAN.

```python
ca_2.drop(['country_code', 'place_name', 'state_name', 'state_code', 'county_name', 'county_code', 'community_name', 'community_code', 'accuracy'], axis = 1, inplace = True)
ca_2['longitude'] = ca_2['longitude'].fillna(-79.6441)
ca_2['latitude'] = ca_2['latitude'].fillna(43.5890)
ca_2
```

### To achieve a common column with dff dataframe, we renamed all the three columns of the ca_2 dataframe, postal_code to Postal Code, latitude to Latitude and longitude to Longitude.

```python
ca_2.rename(columns = {"postal_code": "Postal Code", "latitude": "Latitude", "longitude": "Longitude"}, inplace = True)
ca_2
```

### It is important to check whether the dataframe ca_2 has the same dimension as the dff dataframe. That is, the number of rows must correspond for both dataframes.

```python
ca_2.shape
```

### The next thing to do is to merge the two dataframes in terms of their common column, which is the Postal Code. 

```python
df_new = pd.merge(dff, ca_2, how = 'inner', on = 'Postal Code')
df_new
```
